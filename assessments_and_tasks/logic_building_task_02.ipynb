## Problem 1: Unique Words in a Sentence

```python
# Input sentence
sentence = "python is easy and python is powerful"

# Split sentence into individual words
words = sentence.split()

# Convert list to set to remove duplicate words
unique_words = set(words)

# Display results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

## Problem 2: Highest Salary from Employee Data

```python
# Dictionary storing employee names and salaries
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with maximum salary using dictionary values
highest_employee = max(employees, key=employees.get)

# Display result
print(f"Highest Salary: {highest_employee} - {employees[highest_employee]}")
```

## Problem 3: Find Maximum and Minimum Values

```python
# List of numbers
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = minimum = numbers[0]

# Loop through list to compare values
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Display results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

## Problem 4: Count Products Above a Price Threshold

```python
# List of product prices
prices = [450, 1200, 899, 1500, 300]

count = 0  # Counter for prices above 1000

# Check each price
for price in prices:
    if price > 1000:
        count += 1

# Display result
print("Products above 1000:", count)
```

## Problem 5: Calculate Attendance Percentage

```python
# Attendance list (P = Present, A = Absent)
attendance = ["P", "P", "A", "P", "P"]

# Count present days
present_days = attendance.count("P")

# Calculate attendance percentage
percentage = (present_days / len(attendance)) * 100

# Display result
print("Attendance Percentage:", percentage)
```

## Problem 6: Remove Duplicate Phone Numbers

```python
# List containing duplicate phone numbers
phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert to set to keep only unique numbers
unique_numbers = set(phone_numbers)

# Display unique phone numbers
print("Unique phone numbers:", unique_numbers)
```

## Problem 7: Count Character Frequency

```python
# Input string
text = "python"

# Empty dictionary to store character counts
char_count = {}

# Loop through each character in string
for char in text:
    # Increase count using get() to handle first occurrence
    char_count[char] = char_count.get(char, 0) + 1

# Display result
print(char_count)
```

## Problem 8: Convert List to Tuple

```python
# Original list
lst = [10, 20, 30]

# Convert list to tuple (read-only structure)
tup = tuple(lst)

# Display tuple
print("Tuple:", tup)
```

## Problem 9: Check if a Key Exists in a Dictionary

```python
# Employee dictionary
employees = {"Ravi": 75000, "Anita": 68000}

# Key to search
key = "Ravi"

# Check existence using 'in'
if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

## Problem 10: Calculate Average Marks 

```python
# List of student marks
marks = [70, 75, 80, 65]

# Calculate average using sum and length
average = sum(marks) / len(marks)

# Display result
print("Average Marks:", average)
```
