## Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes):
    total_likes = 0
    
    # calculating total likes using loop
    for value in likes:
        total_likes = total_likes + value

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Sample Run
analyze_engagement([150, 300, 250, 400])
```

## Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def rainfall_checker(rainfall_data, required_level):
    total = 0
    
    # calculating total rainfall
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        result = "Adequate Rainfall"
    else:
        result = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", result)


# Sample Run
rainfall_checker([60, 80, 75, 70], 70)
```

## Problem 4: Social Media – Duplicate Account Detection

```python
def duplicate_detector(usernames):
    
    unique_set = set(usernames)

    if len(unique_set) != len(usernames):
        answer = "Yes"
    else:
        answer = "No"

    print("Duplicate Accounts Found:", answer)


# Sample Run
duplicate_detector(["userA", "userB", "userA", "userC"])
```

## Problem 7: System Monitoring – Application Health Checker

```python
def app_health(errors):

    if errors == 0:
        state = "Healthy"
    elif errors <= 5:
        state = "Minor Issues"
    else:
        state = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", state)


# Sample Run
app_health(7)
```

## Problem 9: E-Learning – Student Attendance Eligibility System

```python
def attendance_system(attendance_record):

    total_days = len(attendance_record)
    present_days = 0

    for day in attendance_record:
        if day == 1:
            present_days += 1

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        result = "Eligible"
    else:
        result = "Not Eligible"

    print("Attendance Percentage:", round(percentage, 1))
    print("Exam Eligibility:", result)


# Sample Run
attendance_system([1, 1, 0, 1, 1])
```

## Problem 10: Smart Electricity Bill Analyzer

```python
def electricity_bill_calculator(units):

    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Level:", usage)


# Sample Run
electricity_bill_calculator(250)
```
