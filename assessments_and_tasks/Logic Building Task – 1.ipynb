# Logic Building Task – 1


## Task 1 – User Login Check


```python
correct_username = "admin"
correct_password = "1234"

user_name = input("Enter username: ")
user_pass = input("Enter password: ")

if user_name == correct_username and user_pass == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

## Task 2 – Pass / Fail Analyzer


```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Pass:", pass_count)
print("Fail:", fail_count)
```

## Task 3 – Simple Data Cleaner


```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned = []

for n in names:
    cleaned.append(n.strip().lower())

print(cleaned)
```

## Task 4 – Message Length Analyzer


```python
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print(msg, "→", length)

    if length > 10:
        print("Long message")
```

## Task 5 – Error Message Detector


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR:", error_count)
```
