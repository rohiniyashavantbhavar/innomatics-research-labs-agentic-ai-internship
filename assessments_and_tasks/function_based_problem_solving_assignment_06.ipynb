## Problem 1: Smart Parking Lot Management System

```python
def smart_parking_system(capacity, vehicle_logs):
    parked = 0
    peak = 0
    
    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
        
        if parked > peak:
            peak = parked
    
    print("Currently Parked Vehicles:", parked)
    
    if parked > capacity:
        print("Parking Status: Full - Over Capacity Alert!")
    else:
        print("Parking Status: Available")
    
    print("Peak Usage:", peak)

# Sample Test
smart_parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

## Problem 2: Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5  # 5 minutes per km
    
    # Traffic delay
    if traffic.lower() == "high":
        base_time += 15
    elif traffic.lower() == "medium":
        base_time += 10
    
    # Weather delay
    if weather.lower() == "rainy":
        base_time += 10
    elif weather.lower() == "storm":
        base_time += 20
    
    print("Estimated Delivery Time:", base_time, "minutes")

# Sample Test
estimate_delivery_time(8, "High", "Rainy")
```

## Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
def theatre_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100
    
    print("Occupancy:", round(occupancy, 2), "%")
    
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")

# Sample Test
theatre_occupancy(200, [1]*150)
```

## Problem 4: Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    average = sum(cpu_readings) / len(cpu_readings)
    
    print("Average CPU Load:", round(average, 2), "%")
    
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

# Sample Test
classify_server_load([45, 60, 70, 85, 90])
```

## Problem 5: Smart Classroom Resource Usage Monitor

```python
def classroom_monitor(resource_usage):
    overused = []
    
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)
    
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

# Sample Test
classroom_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

## Problem 6: Online Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

# Sample Test
event_registration(100, 105)
```
