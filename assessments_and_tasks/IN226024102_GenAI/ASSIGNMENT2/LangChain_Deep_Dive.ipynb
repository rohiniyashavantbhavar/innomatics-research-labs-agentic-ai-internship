## Task 2: Multi-Input Prompt

## Task 5: Input Validation

```python
# Function to validate audience and tone inputs
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Assign default if audience is invalid
    if audience not in valid_audience:
        audience = "beginner"

    # Assign default if tone is invalid
    if tone not in valid_tone:
        tone = "formal"

    return audience, tone
```

```python
# Testing validation function
test_inputs = [
    ("beginner", "fun"),
    ("expert", "formal"),
    ("invalid", "casual"),     # invalid audience
    ("intermediate", "wrong")  # invalid tone
]

for audience, tone in test_inputs:
    validated = validate_inputs(audience, tone)
    print(f"Input: ({audience}, {tone}) → Output: {validated}")
```

```python
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        raise ValueError("Invalid audience")

    if tone not in valid_tone:
        raise ValueError("Invalid tone")

    return audience, tone
```

```python
try:
    result = validate_inputs("beginner", "fun")
    print(result)
except ValueError as e:
    print(e)
```

## Task 6: Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    style = style.lower()  # optional improvement

    if style == "teaching":
        template = "Explain {topic} for {audience} in a {tone} teaching style"

    elif style == "interview":
        template = "Ask interview questions about {topic} for {audience} in a {tone} tone"

    elif style == "storytelling":
        template = "Explain {topic} as a story for {audience} in a {tone} tone"

    else:
        raise ValueError("Invalid style")

    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    return prompt.format(
        topic=topic,
        audience=audience,
        tone=tone
    )
```

```python
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```
