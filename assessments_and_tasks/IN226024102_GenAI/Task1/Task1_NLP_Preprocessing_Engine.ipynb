## Task 1: Conceptual Understanding

**1. Difference between Love and love:**  
Case sensitivity matters; both are treated as different tokens unless normalized.

**2. Without stopwords removal:**  
Adds noise and reduces efficiency.

**3. Harmful cases:**  
- Sentiment analysis  
- Question answering  

**4. Stemming vs Lemmatization:**  
Stemming is rule-based, lemmatization is dictionary-based.


```python
import re
from collections import Counter
```

## Task 2: Preprocessing Function

```python

def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    text = re.sub(r"http\S+|www\S+|@\S+", "", text)
    text = re.sub(r"\d+", "", text)
    text = text.lower()
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()
    tokens = [w for w in tokens if len(w) > 2 or w in ["no", "not"]]

    return tokens, " ".join(tokens)
```

```python

# Task 2 Output
sample_text = "WOW!!! This is soooo AMAZING 😍😍 Visit https://test.com now!!!"
tokens, cleaned = preprocess_text(sample_text)

print("Original Text:", sample_text)
print("Tokens:", tokens)
print("Cleaned Sentence:", cleaned)
```

## Task 3: Stress Testing

```python

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for t in test_sentences:
    tokens, clean = preprocess_text(t)
    print("Original:", t)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    print()
```

## Task 4: Token Analytics

```python

def token_analysis(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg = sum(len(w) for w in tokens)/total if total else 0
    return total, unique, avg

for text in test_sentences:
    tokens, _ = preprocess_text(text)
    total, unique, avg = token_analysis(tokens)

    print("\nText:", text)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg, 2))
```

## Task 5: Frequency Analysis

```python

all_tokens = []
for t in test_sentences:
    tokens, _ = preprocess_text(t)
    all_tokens.extend(tokens)

from collections import Counter
counter = Counter(all_tokens)

print("Top 10 frequent words:")
print(counter.most_common(10))

print("\nLeast frequent words:")
print(counter.most_common()[:-6:-1])
```

## Task 6: Full Pipeline

```python

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    for t in text_list:
        tokens, clean = preprocess_text(t)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)
    return {"tokens": all_tokens, "clean_sentences": clean_sentences}

result = full_pipeline(test_sentences)

print("Tokens:", result["tokens"])
print("\nClean Sentences:", result["clean_sentences"])
```

## Task 7: Error Handling

```python

edge_cases = ["", "😂😂😂", "12345", None]

for e in edge_cases:
    tokens, clean = preprocess_text(e)
    print("Input:", e)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    print()
```
