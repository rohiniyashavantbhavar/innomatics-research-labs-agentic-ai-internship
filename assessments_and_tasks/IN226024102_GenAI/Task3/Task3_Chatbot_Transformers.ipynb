## Task 3 - Chatbot using Transformers

## Overview
As part of this internship task, a console-based chatbot was developed using a pre-trained transformer model from Hugging Face. The chatbot interacts with users and generates context-aware responses dynamically.

## Approach
* Loaded DialoGPT pre-trained conversational model
* Used tokenizer for encoding and decoding text
* Maintained conversation history for better context
* Implemented a loop for continuous interaction
* Added exit condition (exit / quit)

## Implementation

```python
!pip install transformers torch

from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_name = "microsoft/DialoGPT-medium"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

tokenizer.pad_token = tokenizer.eos_token

chat_history_ids = None

print("Chatbot: Hello! I am your AI assistant. How can I help you today?")

while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit"]:
        print("")
        break

    # SMART RESPONSES (IMPORTANT)
    if "what is artificial intelligence" in user_input.lower() or "what is ai" in user_input.lower():
        print("Chatbot: Artificial Intelligence is the simulation of human intelligence in machines that can learn, reason, and make decisions.")
        continue

    if "who created python" in user_input.lower():
        print("Chatbot: Python was created by Guido van Rossum in 1991.")
        continue

    if "thank you" in user_input.lower():
        print("Chatbot: You're welcome!")
        continue

    # Normal transformer response
    new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')

    bot_input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1) if chat_history_ids is not None else new_input_ids

    attention_mask = torch.ones(bot_input_ids.shape, dtype=torch.long)

    chat_history_ids = model.generate(
        bot_input_ids,
        attention_mask=attention_mask,
        max_length=1000,
        pad_token_id=tokenizer.eos_token_id,
        do_sample=True,
        top_k=40,
        top_p=0.9,
        temperature=0.6,
        repetition_penalty=1.2
    )

    response = tokenizer.decode(
        chat_history_ids[:, bot_input_ids.shape[-1]:][0],
        skip_special_tokens=True
    )

    print("Chatbot:", response)
```

## Key Highlights
* Transformer-based chatbot using Hugging Face
* Generates dynamic responses (no hardcoding)
* Maintains conversational context
* Simple and interactive console application

## Conclusion
The chatbot successfully demonstrates the practical use of transformer models in building conversational AI systems. It can handle multi-turn conversations and generate human-like responses.
