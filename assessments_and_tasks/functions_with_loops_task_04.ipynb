## Problem 1 — Mobile Recharge Validation

```python
# Mobile Recharge Validation System
# Takes user input and validates recharge amount

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    amount = int(input("Enter recharge amount: ₹"))

    if amount < 50:
        print("Invalid: Minimum recharge is ₹50.")
    elif amount in valid_plans:
        print("Recharge successful!")
    else:
        print("Invalid plan selected.")


# Run function
validate_recharge()
```

## Problem 2 — Inventory Reorder Alert

```python
# Inventory Reorder Alert System
# Checks stock and prints status

def inventory_alert():
    products = {
        "Pen": int(input("Enter stock for Pen: ")),
        "Notebook": int(input("Enter stock for Notebook: ")),
        "Marker": int(input("Enter stock for Marker: "))
    }

    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")


# Run function
inventory_alert()
```

## Problem 3 — Student Result Processing

```python
# Student Result Processing System
# Calculates average and prints Pass/Fail

def student_result():
    marks = []

    for i in range(3):
        mark = int(input(f"Enter mark {i+1}: "))
        marks.append(mark)

    total = 0
    for m in marks:
        total += m

    average = total / len(marks)

    print(f"Average Marks = {average:.2f}")

    if average >= 50:
        print("Result: Pass")
    else:
        print("Result: Fail")


# Run function
student_result()
```

## Problem 4 — Cab Fare Estimator

```python
# Cab Fare Estimator
# Calculates fare with optional peak hour charge

def calculate_fare():
    BASE_FARE = 50
    PER_KM_RATE = 12

    km = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    fare = BASE_FARE + (PER_KM_RATE * km)

    if peak == "yes":
        fare *= 1.25

    print(f"Total Fare = ₹{fare:.2f}")


# Run function
calculate_fare()
```

## Problem 5 — Attendance Eligibility Checker

```python
# Attendance Eligibility Checker
# Determines eligibility based on attendance percentage

def attendance_eligibility():
    attendance = []

    for i in range(5):
        day = input(f"Enter attendance for day {i+1} (P/A): ").upper()
        attendance.append(day)

    present_days = 0
    for d in attendance:
        if d == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    print(f"Attendance Percentage = {percentage:.2f}%")

    if percentage >= 75:
        print("Status: Eligible")
    else:
        print("Status: Not Eligible")


# Run function
attendance_eligibility()
```

## Problem 6 — Password Strength Checker

```python
# Password Strength Checker
# Checks length, digit, and special character

def check_password_strength():
    password = input("Enter password: ")

    has_digit = False
    has_special = False
    special_chars = "@#$"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if len(password) >= 8 and has_digit and has_special:
        print("Strong Password")
    else:
        print("Weak Password")


# Run function
check_password_strength()
```
